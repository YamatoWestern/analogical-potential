```python
import os
import torchvision
import pandas as pd
```

```python
ds = torchvision.datasets.MNIST(save_dataset_path, download=True)
# ds = torchvision.datasets.FashionMNIST('../dataset/', download=True)
```

```python
dataset = 'MNIST'
# dataset = 'FashionMNIST'

number_of_class = 10

def convert_to_image(path):
    if not os.path.exists(path):
        os.makedirs(path)
    for i in range(number_of_class):
        tpath = path + str(i)
        if not os.path.exists(tpath):
            os.makedirs(tpath)
    for i in range(len(ds)):
        obj, label = ds[i]
        obj.save(path + str(label) + '/' + str(i) + '.png')

path = save_dataset_path+dataset+'/pic/training/'
convert_to_image(path)

ds = torchvision.datasets.MNIST(save_dataset_path, train=False, download=True)
# ds = torchvision.datasets.FashionMNIST('../dataset/', train=False, download=True)

path = save_dataset_path+dataset+'/pic/testing/'
convert_to_image(path)
```
